**📖 Scalars**

```python
temperature = 32

print(temperature)
```

📖**Vectors**

*A vector is an ordered list of numbers* 📋


```python
vector = [5,10,15]

print(vector)
```

📖**Matrices**

*A matrix is a collection of rows and columns* 📊


```python
matrix = [
    [5,6],
    [7,8]
]

print(matrix[0][0])
print(matrix[1][1])
```

📖 **Matrix Shape**

*Matrix shape is represented as:*

*Rows × Columns 📐*


```python
matrix_shape = [
 [1,2],
 [3,4],
 [5,6]
]

rows = len(matrix_shape)
cols = len(matrix_shape)

print(rows)
print(cols)
```

📖**Matrix Addition**

*Matrices can be added only if they have the same shape ➕*

```python
import numpy as np

A = np.array([
    [8,6],
    [4,2]
])

B = np.array([
    [1,2],
    [3,1]
])

print(A + B)
```

📖 **Matrix Subtraction**

*Matrix subtraction is performed element by element ➖*

```python

A = np.array([
    [8,6],
    [4,2]
])

B = np.array([
    [1,2],
    [3,1]
])

print(A - B)
```
